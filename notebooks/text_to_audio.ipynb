# Text to Audio

<b>pyttsx3</b><br>
Free But very generic

```python
import pyttsx3
engine = pyttsx3.init()
# Get voices
voices = engine.getProperty("voices")
print(voices[0].languages)
for v in voices:
    if "en_US" in v.languages:
        print(f"{v.languages} id: {v.id}, name: {v.name}")
```

```python
# engine.setProperty("voice","com.apple.speech.synthesis.voice.Princess")
engine.say("Hello World")
engine.runAndWait()
```

# Google Text-to-Speech

```python
from gtts import gTTS
from playsound3 import playsound

tts = gTTS("Hello World", lang="en")
tts.save("hello.mp3")         # save audio file

playsound("hello.mp3")        # play it
```

<b>Piper</b><br>
Much faster and can download voices from https://huggingface.co/rhasspy/piper-voices/tree/main

Saves the file + Play

```python
import piper
import subprocess
import tempfile, os 

text = "Hello world. Where is Canada?"
voicePath = "../voices/en_US-joe-medium.onnx"
process = subprocess.Popen(
    [
        "piper",
        "--model", voicePath,
        "--output_file", "output.wav",
        "--length-scale", "1.1",
        "--sentence-silence", "0.3",
    ],
    stdin=subprocess.PIPE,
    stdout=subprocess.PIPE,
)

audio, _ = process.communicate(text.encode("utf-8"))

subprocess.run(["afplay", "output.wav"])
```

```python
import piper
import subprocess
import tempfile, os 

text = "Hello world. Where is Canada?"
voicePath = "../voices/en_US-joe-medium.onnx"
piper_proc = subprocess.Popen(
    [
        "piper",
        "--model", voicePath,
        "--output-raw",
        "--length-scale", "1.1",
        "--sentence-silence", "0.3",
    ],
    stdin=subprocess.PIPE,
    stdout=subprocess.PIPE,
)

play_proc = subprocess.Popen(
    [
        "ffplay",
        "-nodisp", "-autoexit",
        "-f", "s16le",      # raw 16-bit signed little-endian
        "-ar", "22050",     # piper's default sample rate (check your model)
        "-i", "pipe:0",
    ],
    stdin=piper_proc.stdout,
)

piper_proc.stdin.write(text.encode("utf-8"))
piper_proc.stdin.close()
play_proc.wait()
```
